```python
df=df.dropna(axis=1,thresh=int(0.25*len(df)))
```

```python
df=df.dropna(axis=0,thresh=int(0.25*len(df.columns)))
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy="median")
impute=SimpleImputer(missing_values=np.nan, strategy="most_frequent")
```

```python
num=df.select_dtypes(include=['int64','float64']).columns
char=df.select_dtypes(include=['object']).columns
```

```python
char
```

```python
df[num]=imputer.fit_transform(df[num])
df[char]=impute.fit_transform(df[char])
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False,drop='first')
encoded=ohe.fit_transform(df[char])
encoded=pd.DataFrame(encoded,columns=ohe.get_feature_names_out(char))
df = pd.concat([df, encoded], axis=1)
df= df.drop(char, axis=1)
```

```python
Q1 = df['SalePrice'].quantile(0.25)
Q3 = df['SalePrice'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['SalePrice'] >= lower_bound) & (df['SalePrice'] <= upper_bound)]
```

```python
df
```

```python
x=df.drop('SalePrice',axis=1)
y=df['SalePrice']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(y, kde=True, bins=100, color='teal')

plt.title(f'Distribution of {y}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(x)
```

```python
df
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=0.95)
X_pca=pca.fit_transform(X_scaled)
```

```python
X_pca=pd.DataFrame(X_pca)
X_pca
```

```python
skn_comp=pca.n_components_
skn_comp
```

```python
sk_ratio=pca.explained_variance_ratio_
sk_ratio
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.3, random_state=42)

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python

plt.figure(figsize=(8, 6))
sns.regplot(x=y_test, y=y_pred, line_kws={"color": "red"})
plt.show()
```

```python

score = model.score(X_test, y_test)
score
```

```python
#PCA without sklearn
```

```python
xt=X_scaled.T
```

```python
c=(1/(len(x)-1))*xt@X_scaled
```

```python
eigenvalues,eigenvectors=np.linalg.eig(c)
```

```python
sorted=eigenvalues.argsort()[::-1]
eigenvalues=eigenvalues[sorted]
eigenvectors=eigenvectors[:,sorted]
```

```python
total_eigenvalue=sum(eigenvalues)
```

```python
varience=eigenvalues/total_eigenvalue;
```

```python
cum_var=np.cumsum(varience)
```

```python
n_comp=np.argmax(cum_var>=0.95)+1
```
